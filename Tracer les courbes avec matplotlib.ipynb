### Exercice

```python
dataset = {f'experience{i}':np.random.randn(100) for i in range(4)}
```

```python
def graphique(data):
    # crÃ©er une fonction qui renvoie le graphe ci-dessus.
    n = len(data)
    plt.figure(figsize=(12,8))
    
    for k,i in zip(data.keys(),range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
graphique(dataset)
```

## Graphique 3D

```python
f = lambda x,y: np.sin(x) + np.cos(x + y)
```

```python
x1 = np.linspace(0,5,100)
y1 = np.linspace(0,5,100)
x1,y1 = np.meshgrid(x,y)
z = f(x1,y1)
```

```python
ax = plt.axes(projection='3d')
ax.scatter(x1,y1,z,cmap='plasma')
```
