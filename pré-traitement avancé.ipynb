```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import SGDClassifier
```

```python
iris = load_iris()
X =  iris.data
y = iris.target
```

```python
selector = SelectFromModel(SGDClassifier(random_state=42),
                           threshold='mean')
selector.fit_transform(X,y)
selector.get_support()
```

```python
selector.estimator_.coef_.mean(axis=0)
```

```python
selector.estimator
```

## Selection recurcive des features

```python
from sklearn.feature_selection import RFE,RFECV
```

```python
help(RFECV)
```

```python
selector = RFE(SGDClassifier(random_state=42),
               step=1,
               n_features_to_select=2
)
selector.fit_transform(X,y)
selector.get_support()
```

```python
selector = RFECV(SGDClassifier(random_state=42),
               step=1,
               min_features_to_select=2,
               cv = 5
)
selector.fit_transform(X,y)
selector.get_support()
```

```python
selector.ranking_
```
